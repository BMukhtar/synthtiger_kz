```python
import collections

root = "../PaddleOCR_Mukhtar/train_data/rec/synthtiger/train_more"

# Read the file
with open(f'{root}/gt.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()

# Separate file paths and words
words = [line.split('\t')[1].strip() for line in lines]

# lower case
words = [word.lower() for word in words]

# Count characters
char_count = collections.Counter()
for word in words:
    char_count.update(word)

# Count words
word_count = collections.Counter(words)

# Display top characters and words
print('Top 10 Characters:')
for char, count in char_count.most_common(50):
    print(f'{char}: {count}')

print('\nTop 10 Words:')
for word, count in word_count.most_common(50):
    print(f'{word}: {count}')
```
